```python
catalognm = 'onedevcatalog'
schemanm = 'gold_one'
```

```python
# Step 1: Create the table by inferring schema from CSV

tablenm = catalognm+'.'+schemanm+'.'+'customer_dim'
create_q = f"""CREATE TABLE {tablenm} \
(
coustomer_key STRING,
name STRING,
contact_no BIGINT,
nid BIGINT,
createdBy STRING,
updatedBy STRING,
createdTimestamp TIMESTAMP,
updatedTimestamp TIMESTAMP
);"""

# Drop the table if it exists
spark.sql(f"DROP TABLE IF EXISTS {tablenm}")
spark.sql(create_q)
```
