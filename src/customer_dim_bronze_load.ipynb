```python
from pyspark.sql.functions import lit,current_timestamp
from datetime import datetime

pathVol=r"/Volumes/onedevcatalog/customer_data/customer-vol/rawfiles/"
cust_dim = pathVol+r'customer_dim.csv'
item_dim = pathVol+r'item_dim.csv'
store_dim = pathVol+r'store_dim.csv'
time_dim = pathVol+r'time_dim.csv'
trans_dim = pathVol+r'Trans_dim.csv'

catalognm = 'onedevcatalog'
schemanm = 'bronze_one'

curr_timestamp = datetime.now()
```

### Load Customer Dimension table

```python


df = spark.read.csv(cust_dim, header=True, inferSchema=True)
print(df.count())
df_upd = df.withColumn("load_time",lit(curr_timestamp))
df_upd = df_upd.withColumn("load_file",lit(cust_dim))
#df_upd.display()
tablenm=f"{catalognm}.{schemanm}.customer_dim"
df_upd.write.mode("overwrite").saveAsTable(tablenm)
```
